```python
import numpy as np
import numpy.linalg as la
import PIL
from PIL import Image, ImageOps
from matplotlib.image import imread
import matplotlib.pyplot as plt
import os

#Part 1 SVD
#1a). Convert the image image1 into a grayscale matrix, gsFull, with values between 0 and 1.
img = Image.open('image1.jpg')
img = ImageOps.grayscale(img) #Converting the image to a grayscale image
gsFull = np.array(img)/255 #Converting the grayscale image to a matrix

#1b). Apply SVD to gsFull to obtain the matrices U,  and V.
U, s, VT = la.svd(gsFull) #SVD of the matrix named gsFull

#1c).Reconstruct the image using U,  and V.
Sigma = np.zeros((gsFull.shape[0],gsFull.shape[1])) #Create a m x n Sigma matrix
Sigma[:gsFull.shape[1], :gsFull.shape[1]] = np.diag(s) #Populate Sigma with n x n diagonal matrix
final_array = U.dot(Sigma.dot(VT)) #reconstructing the matrix
np.allclose(gsFull,final_array)
imgreconstructed = final_array*255
Image.fromarray(imgreconstructed.astype('uint8'))
```

```python
#Part 2 Matrix approximation
#2a). Using SVD, construct a new matrix gsApprox which is a rank 20 approximation of gsFull. Display the approximated image.
U, s, VT = la.svd(gsFull) #SVD of the matrix named gsFull
s = np.diag(s)#extracting the diagonal values
r = 20
gsApprox = np.matrix(U[:,:r] @ s[0:r,:r] @ VT[:r,:])#constructing the approximate image with rank 20)
plt.imshow(gsApprox, cmap = 'gray')
plt.axis('off')
plt.title('rank = ' + str(r))
```

```python
#2b). What is the difference (error) between the original and the approximated image?
mat_diff = np.subtract(gsFull, gsApprox)#first we find the difference between the original and the approximate
ans = la.norm(mat_diff)#then we find the norm of the difference
print("The difference (error) between the original and approximated image is:")
print(ans)
```

```python
#2c). How many elements (numbers) are required to represent the rank-20 matrix, in proportion to the elements used in the original matrix?
U, s, VT = la.svd(gsFull) #SVD of the matrix named gsFull
elements_gsFull = len(U)*len(VT) #Obtaining the elements of the original matrix
elements_gsApprox = (len(U)+len(VT))*r #Obtaining the elements of the approximated matrix
final_percent = (elements_gsApprox/elements_gsFull)*100 #Finding the percentage
print("Percentage of elements required to represent the rank-20 matrix, in proportion to the elements used in the original matrix:")
print(final_percent)
```

```python
#2d). At what rank does the approximation matrix require more elements than the original matrix?
num_of_elements = len(U)*len(VT)#Obtaining the elements of the original matrix

rank = 1 #starting with rank 1
approx_of_elements = 0

while approx_of_elements < num_of_elements: #Comparing the number of elements of the original and the approximate image
    approx_of_elements = (len(U)+len(VT))*rank #Obtaining the elements of the approximated matrix
    rank+=1 #counting the rank
    
print(rank-1) #printing the rank at whihc the approximation matrix requires more elements than the original matrix
```
